```python
import pytesseract
from PIL import Image
from pytesseract import Output

image_path = r"C:\Users\mohas\id-validate-urdu\IDCARD.png"

image = Image.open(image_path)

pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'

result = pytesseract.image_to_data(image, lang='ara', config='--psm 6', output_type=Output.DICT)

extracted_text_list = result['text']

extracted_text_list = [text for text in extracted_text_list if text != '']

extracted_text = ' '.join(extracted_text_list)
print("Extracted text:", extracted_text)
```

```python
import fitz 
from PIL import Image, ImageFilter
import io
import numpy as np

def convert_pdf_to_image(pdf_path):
    zoom_x = 2.0  # horizontal zoom
    zoom_y = 2.0  # vertical zoom
    mat = fitz.Matrix(zoom_x, zoom_y)

    doc = fitz.open(pdf_path)
    page = doc.load_page(0)
    pix = page.get_pixmap(matrix=mat, colorspace='gray')

    img_data = pix.tobytes("png")
    image = Image.open(io.BytesIO(img_data))


    return np.array(image) 

file_path = r"C:\Users\mohas\pdf_invoice_reader_2\OA-111320492.pdf"
image = convert_pdf_to_image(file_path)

from paddleocr import PaddleOCR, draw_ocr
ocr = PaddleOCR(use_gpu=False, lang="ar")
result = ocr.ocr(image)
print(result)
recognized_text = []

# for line in result:
#     text = line
#     print(text)
#     recognized_text.append(text)

# print(recognized_text)
```
